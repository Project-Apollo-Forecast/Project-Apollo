```python
#combine the features from the two dataframes
combine1 = pd.concat([andy, corey[['total_revenue_B','total_flights']]], axis=1)
combine1
```

```python
# Identify the column to be moved
column_to_move = 'total_revenue_B'

# Create a list of column names with the desired order
new_column_order = [column_to_move] + [col for col in combine1.columns if col != column_to_move]

# Reindex the DataFrame with the new column order
df = combine1[new_column_order]
```
